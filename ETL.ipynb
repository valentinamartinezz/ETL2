```python
#cargamos el df que es peliculas para hacer unos cambios 
peliculas = net[net['type']=='Movie']
```

```python
#cargamos el df que es series para hacer unos cambios 
series = net[net['type']=='TV Show']
```

```python
#se normaliza la fecha de peliculas
peliculas['date_added'] = pd.to_datetime(peliculas['date_added'])
```

```python
#se eliminan espacios.
series['date_added']=series['date_added'].str.strip()
#se normaliza la fecha de series.
series['date_added'] = pd.to_datetime(series['date_added'])
```

```python
peliculas['duration']=peliculas['duration'].str.replace('min','')
peliculas['duration']=peliculas['duration'].str.strip()
peliculas['duration']=peliculas['duration'].astype(float)
```

```python
peliculas['duration'].unique()
```

```python
series['duration']=series['duration'].str.replace('Seasons','').replace('1 Season','1').astype(int)
```

```python
series.rename(columns={'duration':'temporadas'})
```

```python
def filmografia(actor:str): #Define la funcion Filmografia que recibe una cadena de texto 
    result1 = peliculas[peliculas['cast'].str.contains(actor,na=False)]#filtra registros que contengan la cadena de texto ingresada(es el filtro q se genera pero para peliculas)
    result2= series[series['cast'].str.contains(actor,na=False)]#filtra registros que contengan la cadena de texto ingresada (es el filtro q se genera pero para series)
    result1 = result1[['title','release_year','type']]#sobreescribir result1 con las columna 'title','release_year' y 'type'
    result2 = result2[['title','release_year','type']]#sobreescribir result2 con las columna 'title','release_year' y 'type'
    result=pd.concat([result1,result2],axis=0).sort_values(by='release_year',ascending = True)#concat=concatenar sort=ordenar 
    result['type'] = result['type'].str.replace('Movie','pelicula').replace('TV Show','Serie')
    return result.rename(columns={'title':'Titulo','release_year':'Año de lanzamiento','type':'Tipo'}).reset_index(drop=True)
```

```python
filmografia('Tom Hanks')
```
